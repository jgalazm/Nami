```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ny = int(1680/4)
nx = int(2400/4)
```

```python
hlast = np.loadtxt('../tsunamilab/tlab2D',delimiter=',',skiprows=1).reshape((ny,nx))
hmax= np.loadtxt('../tsunamilab/tlab2D (1)',delimiter=',',skiprows=1).reshape((ny,nx))
arrival= np.loadtxt('../tsunamilab/tlab2D (2)',delimiter=',',skiprows=1).reshape((ny,nx))
```

```python
k = 3
plt.figure(figsize=(6*k,2*k))
plt.pcolormesh(hmax,vmin=0,vmax=1.0,cmap=plt.cm.jet)
plt.colorbar()
c = plt.contour(arrival/60/60,colors='w',linewidths=0.4)
plt.contour(arrival/60/60,range(50),colors='w',linewidths=0.2)
plt.clabel(c, inline=1, fontsize=10)
ax = plt.gca()
ax.set_aspect(1)
```
